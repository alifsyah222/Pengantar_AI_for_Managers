```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Langkah 1: Menyiapkan direktori dataset
# Misalkan dataset berada di folder "data/kue_indonesia" dengan struktur seperti:
# data/kue_indonesia/
# ├── train/
# │   ├── class_1/
# │   ├── class_2/
# │   └── ...
# └── test/
#     ├── class_1/
#     ├── class_2/
#     └── ...

train_dir = 'dataset_Kue_Indonesia/train'
test_dir = 'dataset_Kue_Indonesia/test'

# Langkah 2: Preprocessing Data menggunakan ImageDataGenerator
# Augmentasi data untuk pelatihan
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,          # Normalisasi gambar
    rotation_range=40,         # Rotasi gambar secara acak
    width_shift_range=0.2,     # Geser gambar secara horizontal
    height_shift_range=0.2,    # Geser gambar secara vertikal
    shear_range=0.2,           # Shear untuk data augmentation
    zoom_range=0.2,            # Zoom pada gambar
    horizontal_flip=True,      # Flip horizontal gambar
    fill_mode='nearest'        # Pengisian gambar setelah transformasi
)

test_datagen = ImageDataGenerator(rescale=1.0/255.0)  # Normalisasi gambar untuk data uji

# Memuat data dengan menggunakan flow_from_directory
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),  # Menyesuaikan ukuran gambar
    batch_size=32,
    class_mode='categorical'  # Untuk klasifikasi multi-kelas
)

validation_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Langkah 3: Membuat model CNN
model = Sequential()

# Layer Convolutional pertama
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D((2, 2)))

# Layer Convolutional kedua
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Layer Convolutional ketiga
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Flatten layer
model.add(Flatten())

# Fully connected layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Dropout untuk mencegah overfitting
model.add(Dense(len(train_generator.class_indices), activation='softmax'))  # Output layer sesuai jumlah kelas

# Langkah 4: Mengkompilasi model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Menampilkan ringkasan model
model.summary()

# Langkah 5: Melatih model
history = model.fit(
    train_generator,
    epochs=5,  # Sesuaikan jumlah epoch sesuai kebutuhan
    validation_data=validation_generator,
    verbose=1
)

# Langkah 6: Evaluasi model
test_loss, test_acc = model.evaluate(validation_generator, verbose=2)
print(f'\nTest accuracy: {test_acc:.2f}')

# Langkah 7: Visualisasi hasil pelatihan
# Plot akurasi
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# Langkah 1: Menyiapkan direktori dataset
train_dir = 'dataset_Kue_Indonesia/train'
test_dir = 'dataset_Kue_Indonesia/test'

# Langkah 2: Preprocessing Data menggunakan ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.8, 1.2]  # Menambahkan range kecerahan
)

test_datagen = ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Langkah 3: Menggunakan Pretrained Model (VGG16) untuk Transfer Learning
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

# Membekukan layer pretrained
base_model.trainable = False

model = Sequential()
model.add(base_model)  # Menambahkan model pretrained
model.add(GlobalAveragePooling2D())  # Global Average Pooling
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_generator.class_indices), activation='softmax'))  # Output layer

# Langkah 4: Mengkompilasi model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Menampilkan ringkasan model
model.summary()

# Langkah 5: Menambahkan EarlyStopping dan ModelCheckpoint
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)

# Langkah 6: Melatih model
history = model.fit(
    train_generator,
    epochs=40,
    validation_data=validation_generator,
    callbacks=[early_stopping, model_checkpoint],
    verbose=1
)

# Langkah 7: Evaluasi model
test_loss, test_acc = model.evaluate(validation_generator, verbose=2)
print(f'\nTest accuracy: {test_acc:.2f}')

# Langkah 8: Visualisasi hasil pelatihan
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
import os
from tensorflow.keras.preprocessing import image
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan prediksi pada gambar input manual dengan label yang benar (otomatis)
def test_manual_image_with_true_label(model, img_path):
    """
    Menguji model dengan gambar manual dan menampilkan label yang benar dan prediksi model.
    
    Args:
        model: Model CNN yang sudah dilatih.
        img_path: Path menuju gambar yang ingin diuji.
    """
    # Memuat gambar dan mengubah ukurannya sesuai dengan input model
    img = Image.open(img_path)
    img = img.resize((150, 150))  # Ukuran gambar sesuai dengan input model
    img_array = np.array(img)  # Mengubah gambar ke array NumPy
    
    # Jika gambar dalam format grayscale, pastikan menjadi RGB
    if img_array.ndim == 2:
        img_array = np.expand_dims(img_array, axis=-1)  # Menambahkan channel
        img_array = np.repeat(img_array, 3, axis=-1)  # Duplikasi channel agar jadi RGB
    
    # Normalisasi gambar
    img_array = img_array / 255.0
    
    # Menambahkan dimensi batch (model menerima input dengan bentuk [batch_size, height, width, channels])
    img_array = np.expand_dims(img_array, axis=0)
    
    # Melakukan prediksi
    prediction = model.predict(img_array)
    
    # Mengambil label yang diprediksi
    predicted_class = np.argmax(prediction, axis=1)
    predicted_label = list(train_generator.class_indices.keys())[predicted_class[0]]
    
    # Mengambil label yang benar berdasarkan nama folder
    true_label = os.path.basename(os.path.dirname(img_path))  # Nama folder adalah label sebenarnya
    
    # Menampilkan gambar dan hasil prediksi
    plt.imshow(img)
    plt.title(f"True: {true_label}, Predicted: {predicted_label}")
    plt.axis('off')  # Menghilangkan axis
    plt.show()

# Contoh penggunaan dengan gambar manual
img_path = 'dataset_Kue_Indonesia/test/kue_kastengel/1.jpg'  # Ganti dengan path gambar yang ingin diuji
test_manual_image_with_true_label(model, img_path)
```
